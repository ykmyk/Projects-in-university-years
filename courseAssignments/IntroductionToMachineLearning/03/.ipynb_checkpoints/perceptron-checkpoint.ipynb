```python
#!/usr/bin/env python3
import argparse

import argparse

import numpy as np
import sklearn.datasets

parser = argparse.ArgumentParser()
parser.add_argument("--data_size", default=100, type=int, help="Data size")
parser.add_argument("--plot", default=False, const=True, nargs="?", type=str, help="Plot the predictions")
parser.add_argument("--recodex", default=False, action="store_true")
parser.add_argument("--seed", default=42, type=int, help="Random seed")
args = parser.parse_args([] if "__file__" not in globals() else None)    
```

```python
generator = np.random.RandomState(args.seed)

# Generate binary classification data with labels {-1, 1}.
data, target = sklearn.datasets.make_classification(
    n_samples=args.data_size, n_features=2, n_informative=2, n_redundant=0,
    n_clusters_per_class=1, flip_y=0, class_sep=2, random_state=args.seed)
target = 2 * target - 1

X = np.c_[data, np.ones(data.shape[0], dtype=int)]
weights = np.zeros(data.shape[1])

done = False
```

```python
while not done:
    permutation = generator.permutation(data.shape[0])
    for i in range(data.shape[0]):
        y = data[i] @ weights
        if(target[i] * y <= 0):
            weights = weights + target[i] @ np.transpose(data[i])
```
